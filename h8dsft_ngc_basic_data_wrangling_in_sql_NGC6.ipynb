```python
from google.colab import auth
from google.cloud import bigquery
import pandas as pd

auth.authenticate_user()
project_id = "datascienceh8"
client = bigquery.Client(project=project_id)
```

How many tables are in the austin_bikeshare dataset?

```python
# put your query here
query = """
SELECT COUNT(*) AS jumlah_table
FROM bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.TABLES
"""
# Set up the query
query_job = client.query(query).to_dataframe()
query_job
```

How many columns in the bikeshare_stations table have STRING data?

```python
# put your query here
query = """
SELECT
  COUNT(*) AS jumlah_column
FROM
  `bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.COLUMNS`
WHERE
  table_name = 'bikeshare_stations'
  AND data_type = 'STRING';
"""
# Set up the query
query_job = client.query(query).to_dataframe()
query_job
```

Retrieve all trips where the bike_id is not null and the duration is greater than 1000 minutes.

```python
# put your query here
query = """
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE bike_id IS NOT NULL
  AND duration_minutes > 1000
"""
# Set up the query
query_job = client.query(query).to_dataframe()
query_job
```

Filter trips where the trip_id is "Walk Up" and the end_station_name  is "Stolen".


```python
# put your query here
query = """
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE
  subscriber_type LIKE "%Walk Up%"
  AND (end_station_name LIKE "%Street" OR end_station_name LIKE "%St.%")
"""
# Set up the query
query_job = client.query(query).to_dataframe()
query_job
```
