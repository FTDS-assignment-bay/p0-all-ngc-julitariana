Create a library catalog system using OOP. The system should allow users to add books to the catalog, search for books by title or author, and remove books from the catalog.
Requirements:
    1.Each book should have a title, author, and unique identifier
    2.The system should provide the following functionality:
        ●Adding a book to the catalog.
        ●Searching for books by title or author.
        ●Removing a book from the catalog.
        ●Displaying all books in the catalog.
    3.Do the exception handling


```python
class Catalog:
    def __init__(self):
        self.catalog = []
    
    def adding_book(self):
        '''Fungsi untuk menambahkan buku ke dalam catalog. Data yang diinputkan adalah id, title, author.'''
        id = input("Input ID: ")
        title = input("Input Title: ")
        author = input("Input Author: ")
        self.catalog.append([id, title, author])
        print("\nThe book's data has been saved!\n")

    def searching_book(self):
        '''Fungsi ini untuk mencari buku berdasarkan judul atau penulis sekaligus.'''
        search = input("Enter the title or author to search for: ").lower()
        found = False
        for book in self.catalog:
            id, title, author = book
            if search in title.lower() or search in author.lower():
                print(f"{id} | {title} | {author}")
                found = True
        if not found:
            print("There are no books with such title nor author")

    def removing_book(self, id):
        '''Fungsi ini untuk menghapus buku dari katalog berdasarkan id yang dimasukkan'''
        for book in self.catalog:
            if book[0] == id:
                self.catalog.remove(book)
                print("\n The book has been removed")
            else: print(f"No book with id {self.id}")

    def display_books(self):
        '''Fungsi ini untuk menampilkan semua buku yang ada di dalam katalog'''
        if not self.catalog:
            print("\n The library catalog system is empty! \n")
        else:
            print("ID | Title | Author")
            for item in self.catalog:
                id, title, author = item
                print(f"{id} | {title} | {author}")

    def main():
        ''''Fungsi utama yang dijalankan oleh sistem'''
        catalog = Catalog()
        while True:
            print("\n---------Welcome to the library catalog system---------")
            print('''Menu
                1. Adding a book to the catalog
                2. Search for books by title or author
                3. Removing a book from the catalog
                4. Display all books in the catalog
                5. Exit''')
            pilih = input("pilih menu: ")
            if pilih == '1':
                print("\n==============================")
                print("Add a book to the library")
                print("==============================\n")
                catalog.adding_book()
                # catalog.display_books()
            elif pilih == '2':
                print("\n===================================")
                print("Search for books by titile or author")
                print("===================================\n")
                catalog.searching_book()
            elif pilih == '3':
                print("\n==============================")
                print("Remove a book from the catalog")
                print("==============================\n")
                id = input("Insert book's id to remove: ")
                catalog.removing_book(id)
            elif pilih == '4':
                print("\n==============================")
                print("Display all books in the catalog")
                print("==============================\n")
                catalog.display_books()
            elif pilih == '5':
                break
    
if __name__ == '__main__':
    '''Pemanggilan fungsi utama dari class Catalog'''
    Catalog.main()
```
